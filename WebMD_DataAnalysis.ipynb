```python
import numpy as np
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import random
import seaborn as sns
```

```python
ADHD_df = pd.read_csv("ADHD.csv",index_col=0)
anxiety_df = pd.read_csv("anxiety.csv",index_col=0)
insomnia_df = pd.read_csv("insomnia.csv",index_col=0)
depression_df = pd.read_csv("depression.csv",index_col=0)
drug_data = [ADHD_df,anxiety_df,insomnia_df,depression_df]
disease = ['ADHD','anxiety','insomnia','depression']
```

```python
def random_color():
        rand = lambda: random.randint(0, 255)
        return '#%02X%02X%02X' % (rand(), rand(), rand())
```

```python
#change indication to numeric and plot pie chart
def indic_plot(data,title):
    data['indication']=data['indication'].astype('category')
    indic = data['indication']
    data['indication']=data['indication'].cat.codes
    plt.title(title)
    plt.pie(indic.value_counts(),labels=indic.value_counts().keys(), colors = [random_color(),random_color()],
            autopct='%.1f%%' )
    plt.savefig((title + '.png'))
```

```python
for i in range(len(drug_data)):
    plt.figure(i)
    indic_plot(drug_data[i],disease[i])

    
    
```
